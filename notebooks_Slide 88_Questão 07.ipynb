##### <center> Representação ilustrativa dos componentes principais do sistema </center> #####


#### Unidade de controle(UC) ####
A UC tem a fução de **determinar instruções**, solicitando qual operação será efetuada no **interior da memória principal**(leitura ou escrita).

#### Controlador da Memória ####
Realiza a **interligação da memória** aos barramentos.
Decodifica o endereço enviado e localiza a célula desejada.


### <center> Registradores: </center> ###

Os registradores constituem **memórias internas ao processador**, e, por sua localização, possuem a **maior velocidade de acesso** ao sistema.

São fabricados com o objetivo de atuar diretamente nas instruções solicitadas pela unidade de controle.

Possuem **capacidade de armazenamento que comporta um dado**, instrução ou palavra por vez, tendo uma **arquitetura vólatil** para isso.

#### RDM: Registrador de Dados da Memória ####
O RDM tem a função de **armazenar a informação que será lida ou escrita**, dependendo da instrução solicitada.

#### REM: Registrador de Endereços da Memória ####
O REM tem a função de **armazenar o endereço**, ou seja, a posição de determinada célula que será acessada durante a operação de leitura ou escrita.

#### Função e Direção: ####
O barramento de controle realiza o papel de **interligação entre a UC e a MP, transportando instruções.**

É um barramento de **fluxo bidirecional**, ou seja, transporta tanto informações enviadas pelo processador, como as instruções de leitura e escrita, quanto informações fornecidas pela memória, como o comando de espera, no caso de operações consecutivas.

```python
Image.open("Image__BD.png")
```

##### <center> Representação ilustrativa do funcionamento do barramento </center> #####

#### Função e Direção: ####
O barramento de Dados **interliga o RDM a MP**, transportando a informação que será lida ou escrita na memória. 

Seu fluxo é **bidirecional**, e é possível determinar qual operação está sendo efetuada por ele. Se a informação está sendo levada para a memória, a operação é escrita, se a informação está sendo levada para o processador, a operação é de leitura.

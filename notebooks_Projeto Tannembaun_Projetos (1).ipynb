#### <center> Para projetar essa memria, primeiramente foi necessrio calcular sua total capacidade, utilizando-se da informao "8Gb"
#### <center> 8 = 2^3
#### <center> GB = 2^30
#### <center> Total = 2^3 * 2^30

```python
Total = 3 + 30
print(f'O total da memria equivale a:2^{Total}')
```

#### <center> Com essa informao, poderemos determinar o restante da memria, pois ela  determinada pelo clculo:
#### <center> 2^33 = Tamanho do endereo * Tamanho da Clula 
#### <center> Para o meu projeto, irei considerar uma matriz quadrada, ou seja, o numero de entradas ser o mesmo para as linhas e colunas, logo: 
#### <center> RAS -> 2^13 e CAS -> 2^13 


```python
RandC = 13 + 13 
print(f'Os endereamentos do RAS e do CAS ocupam um total de {RandC} bits do endereo')
```

#### <center> Com essa desio estabelecida,  possvel determinar o tamanho da clula que ser usada no projeto, respeitando a capacidade total que a memria possui. 
#### <center> Para o meu projeto, minha clula ir armazenar um mximo de 8 bits, logo: 
#### <center> Tamanho da clula -> 2^3

```python
Cell = 3 
RandC = 13 + 13 
t = Cell + RandC 
D = 2**3 + 2**3
print(f'O endereamento da clula ir ocupar um total de {Cell} bits')
print(f'Com isso, ocupamos um total de {t} bits no espao total do endereamento')
print(f'Como o BD ir transportar duas palavras por vez, sero necessrias {D} sadas de dados na memria, logo, o tamanho de uma nica palavra ser 8')
print(f'O tamanho do BD ser igual a: {D} bits')
```

#### <center> Definindo isso, ir sobrar um total de 4 bits para completar o valor estabelecido como total da memria, bits esses que sero destinados aos bancos.

```python
b = 2**4
total = 13 + 13 + 3 + 4
be = 13 + 2 + 4
e = 13 + 13 + 4
print(f'Teremos um total de {b} bancos na memria')
print(f' possvel comprovar todos esses parmetros, pois ao calcular 2^13 * 2^13 * 2^4 * 2^3 iremos atingir o resultado: 2^{total}')
print(f'O tamanho total do barramento de endereos ser: {be}')
print(f'O tamanho total do endereo ser: {e} bits')
```

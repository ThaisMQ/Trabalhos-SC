#### <center> Para projetar essa memória, primeiramente foi necessário calcular sua total capacidade, utilizando-se da informação "8Gb"
#### <center> 8 = 2^3
#### <center> GB = 2^30
#### <center> Total = 2^3 * 2^30

```python
Total = 3 + 30
print(f'O total da memória equivale a:2^{Total}')
```

#### <center> Com essa informação, poderemos determinar o restante da memória, pois ela é determinada pelo cálculo:
#### <center> 2^33 = Tamanho do endereço * Tamanho da Célula 
#### <center> Para o meu projeto, irei considerar uma matriz quadrada, ou seja, o numero de entradas será o mesmo para as linhas e colunas, logo: 
#### <center> RAS -> 2^13 e CAS -> 2^13 


```python
RandC = 13 + 13 
print(f'Os endereçamentos do RAS e do CAS ocupam um total de {RandC} bits do endereço')
```

#### <center> Com essa desição estabelecida, é possível determinar o tamanho da célula que será usada no projeto, respeitando a capacidade total que a memória possui. 
#### <center> Para o meu projeto, minha célula irá armazenar um máximo de 8 bits, logo: 
#### <center> Tamanho da célula -> 2^3

```python
Cell = 3 
RandC = 13 + 13 
t = Cell + RandC 
D = 2**3 + 2**3
print(f'O endereçamento da célula irá ocupar um total de {Cell} bits')
print(f'Com isso, ocupamos um total de {t} bits no espaço total do endereçamento')
print(f'Como o BD irá transportar duas palavras por vez, serão necessárias {D} saídas de dados na memória, logo, o tamanho de uma única palavra será 8')
print(f'O tamanho do BD será igual a: {D} bits')
```

#### <center> Definindo isso, irá sobrar um total de 4 bits para completar o valor estabelecido como total da memória, bits esses que serão destinados aos bancos.

```python
b = 2**4
total = 13 + 13 + 3 + 4
be = 13 + 2 + 16
e = 13 + 13 + 4
print(f'Teremos um total de {b} bancos na memória')
print(f'É possível comprovar todos esses parâmetros, pois ao calcular 2^13 * 2^13 * 2^4 * 2^3 iremos atingir o resultado: 2^{total}')
print(f'O tamanho total do barramento de endereços será: {be}')
print(f'O tamanho total do endereço será: {e} bits')
```

#### <center> Para projetar essa memória, primeiramente foi necessário calcular sua total capacidade, utilizando-se da informação "32Gb"
#### <center> 32 = 2^5
#### <center> Gb = 2^30
#### <center> Total = 2^5 * 2^30

```python
total = 5 + 30
print(f'Essa memória possui um total de armazenamento de 2^{total} bits')
```

#### <center> Com essa informação, poderemos determinar o restante da memória, pois ela é determinada pelo cálculo:
#### <center> 2^33 = Tamanho do endereço * Tamanho da Célula
#### <center> Para o meu projeto, irei considerar uma matriz quadrada, ou seja, o numero de entradas será o mesmo para as linhas e colunas, logo:
#### <center> RAS -> 2^13 e CAS -> 2^13
#### <center> Essa determinação respeita o limite de que os pinos de entrada devem ser menores que 14

```python
RandC = 13 + 13 
print(f'Os endereçamentos do RAS e do CAS ocupam um total de {RandC} bits do endereço')
```

#### <center> Como a questão determina 16 pinos para o BD, irei utilizar uma célula de 16 bits, ou seja, tanto a palavra quanto a célula terão o mesmo tamanho

```python
Cell = 2**4 
RandC = 13 + 13 
t = RandC + 4
print(f'O endereçamento da célula irá ocupar um total de {Cell} bits')
print(f'Com isso, ocupamos um total de {t} bits no espaço total do endereçamento')
print(f'Como o BD terá o mesmo tamanho da célula, ele irá transportar uma palavra de {Cell} bits por vez')
```

#### <center> Definindo isso, irá sobrar um total de 5 bits para completar o valor estabelecido como total da memória, bits esses que serão destinados aos bancos.

```python
b = 2**5
total = 13 + 13 + 4 + 5
be = 13 + 2 + 32
e = 13 + 13 + 5
print(f'Teremos um total de {b} bancos na memória')
print(f'É possível comprovar todos esses parâmetros, pois ao calcular 2^13 * 2^13 * 2^5 * 2^4 iremos atingir o resultado: 2^{total}')
print(f'O tamanho total do barramento de endereços será: {be}')
print(f'O tamanho total do endereço será: {e} bits')
```
